```python
import numpy as np
import pandas as pd
import warnings 
warnings.filterwarnings("ignore")

x_train = pd.read_csv("/kaggle/input/iste-transcend24-hackathon/x_train.csv")
x_train
```

```python
y_train = pd.read_csv('/kaggle/input/iste-transcend24-hackathon/y_train.csv')
y_train
```

```python
x_train.isnull().sum()
```

```python
x_train.describe()
```

```python
categorical_columns = ['Dietary Habits Score','Physical Activity Level','Family History of Lung Cancer']
```

```python
num_columns = ['Age','Smoking History','Genetic Predisposition Score','Exposure to Carcinogens','Air Pollution Index','BMI','Income Level']
```

```python
y_train = y_train['Risk'].to_list()
y_train = np.array(y_train)
len(y_train)
```

```python
x_train = pd.get_dummies(x_train, columns=categorical_columns)
x_train
```

```python
x_train.replace({False: 0, True: 1}, inplace=True)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train[num_columns] = scaler.fit_transform(x_train[num_columns])
x_train
```

```python
x_train.drop(['Id'],axis=1, inplace =True)
x_train
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

pca = PCA()
x_train_pca = pca.fit_transform(x_train)
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.title('Explained Variance vs. Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

```python
pca = PCA(n_components=19)
x_train = pca.fit_transform(x_train)
```

```python
x_test = pd.read_csv("/kaggle/input/iste-transcend24-hackathon/test.csv")
x_test.drop(['Id'], axis =1 ,inplace = True)
x_test = pd.get_dummies(x_test, columns=categorical_columns)
x_test[num_columns] = scaler.transform(x_test[num_columns])
x_test.replace({False: 0, True: 1}, inplace=True)
x_test = pca.transform(x_test)
x_test
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor

X_train, X_test, Y_train, Y_test = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train, Y_train)

random_forest_model = RandomForestRegressor()
random_forest_model.fit(X_train, Y_train)

gradient_boosting_model = GradientBoostingRegressor()
gradient_boosting_model.fit(X_train, Y_train)

svr_model = SVR()
svr_model.fit(X_train, Y_train)

knn_model = KNeighborsRegressor()
knn_model.fit(X_train, Y_train)

decision_tree_model = DecisionTreeRegressor()
decision_tree_model.fit(X_train, Y_train)


models = {
    'Linear Regression': linear_reg_model,
    'Random Forest': random_forest_model,
    'Gradient Boosting': gradient_boosting_model,
    'SVR': svr_model,
    'KNN': knn_model,
    'Decision Tree': decision_tree_model
}

for name, model in models.items():
    y_pred = model.predict(X_test)
    mse = mean_squared_error(Y_test, y_pred)
    print(f"{name} Mean Squared Error: {mse}")
```

```python
y_test_predict = gradient_boosting_model.predict(x_test)
result = pd.DataFrame({'Risk': y_test_predict}, index=range(1401, 1550))
result.to_csv('gb.csv', index_label='Id')
```
