```python
temp = pd.read_csv('/kaggle/input/iste-transcend24-hackathon/x_train.csv')
temp_y = pd.read_csv('/kaggle/input/iste-transcend24-hackathon/y_train.csv')
```

```python
X = temp
y = temp_y.Risk

score=[]
for feature in X.columns:
    correlation=temp[feature].corr(y)
    score.append(correlation)
    
correlation_dict = {column: correlation for column, correlation in zip(X.columns, score)}
st = dict(sorted(correlation_dict.items(), key=lambda X: X[1], reverse=True))

my_dict = {"Feature" : [], "Score" : []}
key = [i for i in st.keys()]
score = [i for i in st.values()]
for i in key:
    my_dict['Feature'].append(i)
for i in score:
    my_dict['Score'].append(i)
    
res = pd.DataFrame.from_dict(my_dict)
res
```

```python
x_test = pd.read_csv("/kaggle/input/iste-transcend24-hackathon/test.csv")
x_test.drop(['Id'], axis =1 ,inplace = True)
# x_test = pd.get_dummies(x_test, columns=categorical_columns)
x_test[num_columns] = scaler.transform(x_test[num_columns])
# x_test.replace({False: 0, True: 1}, inplace=True)
# x_test = pca.transform(x_test)
x_test
```

```python
## only smoiking history is important
x_train = np.array(x_train).reshape(-1,1)
x_train
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor

X_train, X_test, Y_train, Y_test = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train, Y_train)

random_forest_model = RandomForestRegressor()
random_forest_model.fit(X_train, Y_train)

gradient_boosting_model = GradientBoostingRegressor()
gradient_boosting_model.fit(X_train, Y_train)

svr_model = SVR()
svr_model.fit(X_train, Y_train)

knn_model = KNeighborsRegressor()
knn_model.fit(X_train, Y_train)

decision_tree_model = DecisionTreeRegressor()
decision_tree_model.fit(X_train, Y_train)


models = {
    'Linear Regression': linear_reg_model,
    'Random Forest': random_forest_model,
    'Gradient Boosting': gradient_boosting_model,
    'SVR': svr_model,
    'KNN': knn_model,
    'Decision Tree': decision_tree_model
}

for name, model in models.items():
    y_pred = model.predict(X_test)
    mse = mean_squared_error(Y_test, y_pred)
    print(f"{name} Mean Squared Error: {mse}")
```

```python
y_test_predict = gradient_boosting_model.predict(x_test)
result = pd.DataFrame({'Risk': y_test_predict}, index=range(1401, 1550))
result.to_csv('gb.csv', index_label='Id')
```
