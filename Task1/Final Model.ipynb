```python
import numpy as np
import pandas as pd
import warnings 
warnings.filterwarnings("ignore")

x_train = pd.read_csv("/kaggle/input/iste-transcend24-hackathon/x_train.csv")
x_train
```

```python
categorical_columns = ['Dietary Habits Score','Physical Activity Level','Family History of Lung Cancer']
```

```python
num_columns = ['Age','Smoking History','Genetic Predisposition Score','Exposure to Carcinogens','Air Pollution Index','BMI','Income Level']
```

```python
y_train = y_train['Risk'].to_list()
y_train = np.array(y_train)
len(y_train)
```

### One hot encode the categorical columns / Label encoding can be done as well

```python
x_train = pd.get_dummies(x_train, columns=categorical_columns)
x_train
```

```python
x_train.replace({False: 0, True: 1}, inplace=True)
x_train
```

#### Scaling to be done as Age has higher values as compared to other features. 
We dont want to add bias to the model, we could have gone with Standard Scaling as well but that would take some of the age values to have negative values, which gives a wrong learning parameter to the model

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train[num_columns] = scaler.fit_transform(x_train[num_columns])
x_train
```

```python
x_train.drop(['Id'],axis=1, inplace =True)
x_train
```

```python
x_test = pd.read_csv("/kaggle/input/iste-transcend24-hackathon/test.csv")
x_test.drop(['Id'], axis =1 ,inplace = True)
x_test = pd.get_dummies(x_test, columns=categorical_columns)
x_test[num_columns] = scaler.transform(x_test[num_columns])
x_test.replace({False: 0, True: 1}, inplace=True)
x_test
```

### Testing the base variant of few models

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor

X_train, X_test, Y_train, Y_test = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train, Y_train)

random_forest_model = RandomForestRegressor()
random_forest_model.fit(X_train, Y_train)

gradient_boosting_model = GradientBoostingRegressor()
gradient_boosting_model.fit(X_train, Y_train)

svr_model = SVR()
svr_model.fit(X_train, Y_train)

knn_model = KNeighborsRegressor()
knn_model.fit(X_train, Y_train)

decision_tree_model = DecisionTreeRegressor()
decision_tree_model.fit(X_train, Y_train)


models = {
    'Linear Regression': linear_reg_model,
    'Random Forest': random_forest_model,
    'Gradient Boosting': gradient_boosting_model,
    'SVR': svr_model,
    'KNN': knn_model,
    'Decision Tree': decision_tree_model
}

for name, model in models.items():
    y_pred = model.predict(X_test)
    mse = mean_squared_error(Y_test, y_pred)
    print(f"{name} Mean Squared Error: {mse}")
```

### HP Tuning of the regressor models that work on ML algorithms and statistics

```python
def find_best_model(X_train, y_train, X_test, y_test):
    best_model = None
    best_r2 = 0.0
    models = {
        'Linear Regression': (LinearRegression(), {
        'fit_intercept': [True, False],
        'copy_X': [True, False],
        'n_jobs': [None, -1]
    }),
    'Random Forest': (RandomForestRegressor(), {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }),
    'Decision Tree': (DecisionTreeRegressor(), {
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }),
    'Gradient Boosting': (GradientBoostingRegressor(), {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.001, 0.01, 0.1, 1],
        'max_depth': [3, 5, 7]
    }),
    'Ada Boost': (AdaBoostRegressor(), {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.001, 0.01, 0.1, 1]
    }),
    'SVM': (SVR(), {
        'C': [0.001, 0.01, 0.1, 1, 10],
        'kernel': ['linear', 'rbf'],
        'gamma': ['scale', 'auto']
    }),
    'KNN': (KNeighborsRegressor(), {
        'n_neighbors': [3, 5, 7],
        'weights': ['uniform', 'distance'],
        'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
    }),
    'Extra Gradient': (ExtraTreesRegressor(), {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }),
    'Bagging': (BaggingRegressor(), {
        'n_estimators': [50, 100, 200],
        'max_samples': [1.0, 0.8, 0.6],
        'max_features': [1.0, 0.8, 0.6]
    }),
    'XGBoost': (XGBRegressor(), {
        'n_estimators': [50, 100, 200, 500],
        'learning_rate': [0.001, 0.01, 0.1, 1],
        'max_depth': [3, 5, 7, 10],
        'gamma': [0, 0.1, 0.2, 0.3]
    }),
    'CatBoost': (CatBoostRegressor(verbose=False), {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.001, 0.01, 0.1, 1],
        'depth': [4, 6, 8, 10]
    })
        
    }

    for name, (model, params) in models.items():
        try:
            grid_search = GridSearchCV(model, params, cv=5, scoring='r2')
            grid_search.fit(X_train, y_train)
            y_pred_cv = grid_search.best_estimator_.predict(X_train)
            mse_cv = mean_squared_error(y_train, y_pred_cv)

            y_pred_test = grid_search.best_estimator_.predict(X_test)
            mse_test = mean_squared_error(y_test, y_pred_test)

            r2 = grid_search.best_score_

            print("Model:", name)
            print("Cross-validation MSE:", mse_cv)
            print("Test MSE:", mse_test)
            print("Cross-validation R-squared:", r2)
            print("\n_________________________________________________________________________________\n")

            if r2 > best_r2:
                best_r2 = r2
                best_model_name = name
                best_model = grid_search.best_estimator_
        except Exception as e:
            print(f"Error for model {name}: {e}")

    return best_model, best_r2
```

```python
best_model, test_accuracy = find_best_model(X_train, Y_train, X_test, Y_test)
print("Best Model:", best_model)
print('Test R2 of the Best Model:', test_accuracy * 100)
```
